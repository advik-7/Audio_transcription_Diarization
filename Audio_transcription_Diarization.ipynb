```python
import whisper
from resemblyzer import VoiceEncoder, preprocess_wav
from sklearn.cluster import SpectralClustering
import numpy as np
import librosa

audio_path = "/content/clipped_audio.mp3"
model = whisper.load_model("base")
transcription_result = model.transcribe(audio_path)
segments = transcription_result["segments"]

wav, sr = librosa.load(audio_path, sr=None)
encoder = VoiceEncoder()

segment_duration = 1.5
step_size = int(sr * segment_duration)
embeddings = []
timestamps = []

for start in range(0, len(wav), step_size):
    end = min(start + step_size, len(wav))
    segment = wav[start:end]
    preprocessed = preprocess_wav(segment, source_sr=sr)
    if len(preprocessed) > 0:
        embedding = encoder.embed_utterance(preprocessed)
        embeddings.append(embedding)
        timestamps.append((start / sr, end / sr))

embeddings = np.array(embeddings)

n_speakers = 2
clustering = SpectralClustering(n_clusters=n_speakers, affinity="nearest_neighbors", assign_labels="kmeans")
labels = clustering.fit_predict(embeddings)

conversation = []

for segment in segments:
    whisper_start = segment["start"]
    whisper_end = segment["end"]
    text = segment["text"]

    speaker_label = None
    for i, (start, end) in enumerate(timestamps):
        if whisper_start < end and whisper_end > start:
            speaker_label = f"Speaker {labels[i] + 1}"
            break

    if speaker_label:
        conversation.append((speaker_label, text))
    else:
        conversation.append(("Unknown Speaker", text))

print("\n--- Transcribed Conversation ---\n")
for speaker, text in conversation:
    print(f"{speaker}: {text}")
```

```python
import whisper
from resemblyzer import VoiceEncoder, preprocess_wav
from sklearn.cluster import SpectralClustering
import numpy as np
import librosa

def transcribe_and_diarize(audio_path, segment_duration=1.5, n_speakers=2):
    model = whisper.load_model("base")
    transcription_result = model.transcribe(audio_path)
    segments = transcription_result["segments"]

    wav, sr = librosa.load(audio_path, sr=None)
    encoder = VoiceEncoder()

    step_size = int(sr * segment_duration)
    embeddings = []
    timestamps = []

    for start in range(0, len(wav), step_size):
        end = min(start + step_size, len(wav))
        segment = wav[start:end]
        preprocessed = preprocess_wav(segment, source_sr=sr)
        if len(preprocessed) > 0:
            embedding = encoder.embed_utterance(preprocessed)
            embeddings.append(embedding)
            timestamps.append((start / sr, end / sr))

    embeddings = np.array(embeddings)

    clustering = SpectralClustering(n_clusters=n_speakers, affinity="nearest_neighbors", assign_labels="kmeans")
    labels = clustering.fit_predict(embeddings)

    conversation = []

    for segment in segments:
        whisper_start = segment["start"]
        whisper_end = segment["end"]
        text = segment["text"]

        speaker_label = None
        for i, (start, end) in enumerate(timestamps):
            if whisper_start < end and whisper_end > start:
                speaker_label = f"Speaker {labels[i] + 1}"
                break

        if speaker_label:
            conversation.append((speaker_label, text))
        else:
            conversation.append(("Unknown Speaker", text))

    return conversation
```
